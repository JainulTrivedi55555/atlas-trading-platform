```python
import pandas as pd
import numpy as np
import sys, warnings
warnings.filterwarnings('ignore')
sys.path.append('../..')
from src.utils.config import TICKERS, PROCESSED_DIR
```

```python
print('=' * 60)
print('ATLAS — Phase 3 EDA Summary')
print('=' * 60)
```

```python
# Dataset Summary 
print('\n1. DATASET OVERVIEW')
print('-' * 40)
for ticker in TICKERS:
    df = pd.read_csv(PROCESSED_DIR / f'features/{ticker}_features.csv',
                     index_col=0, parse_dates=True)
    print(f'  {ticker}: {len(df)} rows, {len(df.columns)} features')
```

```python
#  Return Statistics 
print('\n2. RETURN STATISTICS')
print('-' * 40)
for ticker in TICKERS:
    df = pd.read_csv(PROCESSED_DIR / f'features/{ticker}_clean.csv',
                                       index_col=0, parse_dates=True)
    r = df['Daily_Return'].dropna()
    annual_ret = ((1 + r).prod() ** (252/len(r)) - 1) * 100
    annual_vol = r.std() * np.sqrt(252) * 100
    sharpe = annual_ret / annual_vol
    print(f'  {ticker}: Ann.Ret={annual_ret:.1f}% '
          f'Vol={annual_vol:.1f}% Sharpe={sharpe:.2f}')
```

```python
#  Feature Summary 
print('\n3. FEATURE ENGINEERING SUMMARY')
print('-' * 40)
df = pd.read_csv(PROCESSED_DIR / 'features/AAPL_features.csv',
                 index_col=0, parse_dates=True)
feature_groups = {
    'Price/Return': ['Daily_Return','Log_Return','Return_5d','Return_20d'],
    'Momentum':     ['RSI_14','MACD','Stoch_K','Williams_R','ROC_20'],
    'Trend':        ['SMA_20','SMA_50','SMA_200','Price_vs_SMA200'],
    'Volatility':   ['ATR_14','BB_Width','Volatility_5d','Volatility_20d'],
    'Volume':       ['OBV','Volume_Ratio','CMF'],
}
for group, feats in feature_groups.items():
    present = [f for f in feats if f in df.columns]
    print(f'  {group}: {len(present)} features')
```

```python
# Target Summary 
print('\n4. TARGET VARIABLE SUMMARY')
print('-' * 40)
for ticker in TICKERS:
    y = pd.read_csv(PROCESSED_DIR / f'splits/{ticker}/y_train.csv',
                    index_col=0).squeeze()
    print(f'  {ticker}: {y.mean()*100:.1f}% up days '
        f'({y.sum()} up / {(y==0).sum()} down)')
```
